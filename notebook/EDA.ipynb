```python
##segregate numerical and categorial column

numerical_columns=df.columns[df.dtypes!='object']
categorial_columns=df.columns[df.dtypes=='object']
print("Numerical Columns: ",numerical_columns)
print("Categorial columns: ",categorial_columns)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
x=1

for i in numerical_columns:
    sns.histplot(data=df,x=i,kde=True)
    print('\n')
    plt.show()
```

```python
##Assignment to plot graphs for categorial data
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
x=1

for i in categorial_columns:
    sns.histplot(data=df,x=i,kde=True)
    print('\n')
    plt.show()
```

```python
##Assignment to plot graphs for categorial data
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
x=1

for i in categorial_columns:
    sns.displot(data=df,x=i,kde=True)
    print('\n')
    plt.show()
```
